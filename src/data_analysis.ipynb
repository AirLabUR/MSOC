# Video Analysis

```python
def process_video_data(data):
    dir_split = data.split("/")
    subject1 = dir_split[-2]
    video_split = dir_split[-1].split("-")
    video_name = video_split[0]
    model_name = video_split[1] + "_" + video_split[2]
    bledning_name = video_split[3]
    subject2 = video_split[5].split(".")[0]
    return subject1, subject2, video_name, model_name, bledning_name
```

```python
data_processed = [
    {
        "subject1": subject1, 
        "subject2": subject2, 
        "video_key": video_name, 
        "model_name": model_name, 
        "bledning_name": bledning_name
    }
    for subject1, subject2, video_name, model_name, bledning_name in map(process_video_data, video_data)
]

# Create DataFrame from the processed data
video_df = pd.DataFrame(data_processed, columns=["subject1", "subject2", "video_key", "model_name", "bledning_name"])
```

```python
def process_audio_data(data):
    dir_split = data.split("/")
    dir_split
    subject1 = dir_split[-2]
    audio_split = dir_split[-1].split("-")
    audio_name = audio_split[0]
    audio_model = audio_split[1] + "-" + audio_split[2] + "-" + audio_split[3]

    subject2 = audio_split[-1].split(".")[0]
    return subject1, subject2, audio_name, audio_model
```
